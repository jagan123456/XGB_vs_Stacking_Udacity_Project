# Data Preprocessing

```python
raw.head()
```

## Remove Unnecessary Columns

```python
clean_df = raw.drop(['instance weight','year','veterans benefits','fill inc questionnaire for veteran\'s admin'\
                    ,'industry code','occupation code','own business or self employed'],axis=1)
```

## Cleanse Target Feature

```python
clean_df['target'][clean_df.target == '-50000'] = 0
clean_df['target'][clean_df.target.str.strip() == '50000+.'] = 1
```

```python
clean_df.target = clean_df.target.astype(bool)
```

## Collapse Education Feature

```python
clean_df['education'][clean_df.education.str.strip().isin(['Less than 1st grade','1st 2nd 3rd or 4th grade',\
                                                                   '5th or 6th grade','7th and 8th grade'])] = 'below 9th grade'

clean_df['education'][clean_df.education.str.strip().isin(['9th grade','10th grade',\
                                                                   '11th grade','12th grade no diploma'])] = '9-12th grade (no diploma)'
```

## Standardise Null Values

```python
for feature in clean_df.select_dtypes(['object']):
    clean_df[feature] = clean_df[feature].str.strip()
```

```python
clean_df.replace(['?','Not in universe'],np.NAN,inplace=True)
```

```python
clean_df.head()
```

## Remove Duplicate Rows

```python
clean_df.drop_duplicates(inplace=True)
```

```python
clean_df.shape
```

```python
print( str(raw.shape[0]-clean_df.shape[0]) + " duplicates removed")
```

```python
print ("Proportion of people earning < $50K: %0.2f%%" % (100*clean_df['target'].value_counts()[0]/clean_df.shape[0]))
print ("Proportion of people earning > $50K: %0.2f%%" % (100*clean_df['target'].value_counts()[1]/clean_df.shape[0]))
```

```python
ratio_neg_pos = clean_df['target'].value_counts()[0]/clean_df['target'].value_counts()[1]
print (ratio_neg_pos)
```

## OHE of Categorical Variables

```python
clean_df['detailed household and family stat'] = clean_df['detailed household and family stat'].str.replace('<','less than ')
```

```python
cat_vars = pd.get_dummies(clean_df.select_dtypes('object'))
```

```python
num_vars = clean_df.select_dtypes(exclude ='object')
```

```python
clean_dummy_df = num_vars.merge(cat_vars,left_index=True, right_index=True)
```

## OHE Test

```python
X = clean_dummy_df.drop('target',axis=1)
```

```python
Y = clean_dummy_df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
baseline_model = xgb.XGBClassifier(n_jobs=-1, random_state=1, eval_metric='auc', \
                                       scale_pos_weight=ratio_neg_pos)
```

```python
start = time.time()
print ('Started training: '+ str(strftime("%H:%M:%S", gmtime())))

baseline_model.fit(X_train,y_train)

duration = time.time() - start
print ('Finished training, time taken: %0.2f seconds' % duration)
```

```python
y_pred = baseline_model.predict(X_test)
```

```python
print ('ROC Score: %0.4f' % roc_auc_score(y_test,y_pred))
```

## Encoding of Categorical Variables Only (No OHE) Test

```python
#### This sub-section is used to test performance of the XGBoost model with no OHE on the categorical features.
#### The categorical variables are numerically encoded only. Since a lower AUC was output the OHE variables were 
#### used in sub-sections going forward.
```

```python
cat_vars_enc = clean_df.select_dtypes('object').astype(str)
cat_vars_enc = cat_vars_enc.apply(LabelEncoder().fit_transform)
```

```python
clean_dummy_df_enc = num_vars.merge(cat_vars_enc,left_index=True, right_index=True)
```

```python
X_enc = clean_dummy_df_enc.drop('target',axis=1)
```

```python
Y_enc = clean_dummy_df_enc['target']
```

```python
X_train_enc, X_test_enc, y_train_enc, y_test_enc = train_test_split(X_enc, Y_enc, test_size=0.2, random_state=1)
```

```python
baseline_model_enc = xgb.XGBClassifier(n_jobs=-1, random_state=1, eval_metric='auc', \
                                       scale_pos_weight=ratio_neg_pos)
```

```python
start = time.time()
print ('Started training: '+ str(strftime("%H:%M:%S", gmtime())))

baseline_model_enc.fit(X_train_enc,y_train_enc)

duration = time.time() - start
print ('Finished training, time taken: %0.2f seconds' % duration)
```

```python
y_pred_enc = baseline_model_enc.predict(X_test_enc)
```

```python
print ('ROC Score: %0.4f' % roc_auc_score(y_test_enc,y_pred_enc))
```

# Implementation

## Split No OHE Data Further

```python
X_train_L1M, X_test_L2M, y_train_L1M, y_test_L2M = train_test_split(X_enc, Y_enc, test_size=0.4, random_state=1)
```

```python
X_train_S1, X_test_S1, y_train_S1, y_test_S1 = train_test_split(X_train_L1M, y_train_L1M, test_size=0.2, random_state=1)
```

## Random Forest

```python
start = time.time()
print ('Started training: '+ str(strftime("%H:%M:%S", gmtime())))

param_grid = {'max_features':np.arange(1, 20, 2), 'min_samples_leaf':np.arange(5, 31, 5)}

RF_grid_clf = RandomizedSearchCV(RandomForestClassifier(n_jobs=-1, random_state=1, n_estimators = 800, class_weight='balanced'), \
                                 param_grid, scoring='roc_auc', cv=3, n_jobs=-1, random_state=1)

RF_grid_clf.fit(X_train_S1, y_train_S1)

duration = time.time() - start
print ('Finished training, time taken: %0.2f seconds' % duration)
```

```python
RF_grid_preds = RF_grid_clf.predict(X_test_S1)
RF_grid_performance = roc_auc_score(y_test_S1, RF_grid_preds)
```

```python
print ("Optimum RF parameters from randomizedsearch: " + str(RF_grid_clf.best_params_))
```

```python
print ("ROC performance of the optimum RF model on the test set from randomizedsearch: %0.8f" % RF_grid_performance)
```

```python
optimized_RF_model = RandomForestClassifier(n_jobs=-1, random_state=1, n_estimators=800, class_weight='balanced', \
                                            max_features=RF_grid_clf.best_params_.get('max_features'),\
                                            min_samples_leaf=RF_grid_clf.best_params_.get('min_samples_leaf'))
```

```python
start = time.time()
print ('Started training: '+ str(strftime("%H:%M:%S", gmtime())))

optimized_RF_model.fit(X_train_S1, y_train_S1)

duration = time.time() - start
print ('Finished training, time taken: %0.2f seconds' % duration)
```

```python
y_pred_RF = optimized_RF_model.predict(X_test_S1)
```

```python
print ('ROC Score: %0.4f' % roc_auc_score(y_test_S1, y_pred_RF))
```

## SVM

## Data Scaling

```python
X_enc_scaled = preprocessing.scale(X_enc)
```

```python
X_enc_scaled = pd.DataFrame(X_enc_scaled, columns=X_enc.columns)
```

```python
X_train_SVM_L1M, X_test_SVM_L2M, y_train_SVM_L1M, y_test_SVM_L2M = train_test_split(X_enc_scaled, Y_enc, \
                                                                                    test_size=0.4, random_state=1)
```

```python
X_train_SVM_S1, X_test_SVM_S1, y_train_SVM_S1, y_test_SVM_S1 = train_test_split(X_train_SVM_L1M, y_train_SVM_L1M,\
                                                                                test_size=0.2, random_state=1)
```

## GridSearch

```python
start = time.time()
print ('Started training: '+ str(strftime("%H:%M:%S", gmtime())))
 
param_grid = {'gamma': np.logspace(-2, 2, 2), 'C': np.logspace(-2, 2, 2)}

SVM_grid_clf = GridSearchCV(svm.SVC(cache_size=1000, random_state=1, class_weight='balanced'), \
                                 param_grid, scoring='roc_auc', cv=3, n_jobs=-1)

SVM_grid_clf.fit(X_train_SVM_S1, y_train_SVM_S1)

duration = time.time() - start
print ('Finished training, time taken: %0.2f seconds' % duration)
```

```python
print ("Optimum SVM parameters from randomizedsearch: " + str(SVM_grid_clf.best_params_))
```

```python
print ("ROC performance of the optimum SVM model on the test set from randomizedsearch: %0.8f" % SVM_grid_performance)
```

```python
optimized_SVM_model = svm.SVC(cache_size=1000, random_state=1, class_weight='balanced', C=SVM_grid_clf.best_params_.get('C'), \
                              gamma= SVM_grid_clf.best_params_.get('gamma'))
```

```python
start = time.time()
print ('Started training: '+ str(strftime("%H:%M:%S", gmtime())))

optimized_SVM_model.fit(X_train_SVM_S1, y_train_SVM_S1)

duration = time.time() - start
print ('Finished training, time taken: %0.2f seconds' % duration)
```

```python
y_pred_SVM = optimized_SVM_model.predict(X_test_SVM_S1)
```

```python
print ('ROC Score: %0.4f' % roc_auc_score(y_test_SVM_S1, y_pred_SVM))
```

# Stacked Ensemble

## Merge Different Data Treatments

```python
# need to join the scaled data onto the encoded data
X_test_L2M_combi = X_test_L2M.reset_index(drop=True).merge(X_test_SVM_L2M.reset_index(drop=True), right_index=True, \
                                                           left_index=True, suffixes = ('_enc','_scale')) 
```

```python
# XGB, LR and RF use encoded data
# SVM uses scaled
X_train_S2, X_test_S2, y_train_S2, y_test_S2 = train_test_split(X_test_L2M_combi, y_test_L2M, test_size=0.2, random_state=1)
```

## Stacking Experimentation

```python
pipe1 = make_pipeline(ColumnSelector(cols=(np.arange(34))), optimized_RF_model)
pipe2 = make_pipeline(ColumnSelector(cols=(np.arange(34,68))), optimized_SVM_model)
pipe3 = make_pipeline(ColumnSelector(cols=(np.arange(34))), optimized_LR_model)
pipe4 = make_pipeline(ColumnSelector(cols=(np.arange(34))), optimized_XGB_model)

#meta_classifier=xgb.XGBClassifier(n_jobs=-1,random_state=1,eval_metric='auc', scale_pos_weight=ratio_neg_pos)
#meta_classifier=svm.SVC(cache_size=1000, random_state=1, class_weight='balanced')
meta_classifier=logit(n_jobs=-1,random_state=1, class_weight='balanced')
#meta_classifier=RandomForestClassifier(n_jobs=-1, random_state=1, class_weight='balanced')

sclf = StackingClassifier(classifiers=[pipe1, pipe3, pipe4], meta_classifier=meta_classifier)
```

```python
start = time.time()
print ('Started training: '+ str(strftime("%H:%M:%S", gmtime())))

sclf.fit(X_train_S2.values, y_train_S2)

duration = time.time() - start
print ('Finished training, time taken: %0.2f seconds' % duration)
```

```python
y_pred_sclf = sclf.predict(X_test_S2.values)
```

```python
print ('ROC Score: %0.8f' % roc_auc_score(y_test_S2, y_pred_sclf))
```

# Free-Form Visualization

```python
base_clf_preds = pd.DataFrame([y_pred_RF,y_pred_SVM,y_pred_LR,y_pred_xgb]).transpose()
base_clf_preds.columns = ['Random Forest','SVM','Logistic Regression','XGB']
```

```python
sns.set(style="white")

# Compute the correlation matrix
corr = base_clf_preds.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(8, 16))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(30, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
g = sns.heatmap(corr, mask=mask, annot=True, cmap=cmap, vmax=.8, center=0.55,
            square=True,  cbar_kws={"shrink": .25}, xticklabels='auto', yticklabels='auto')

g.set_yticklabels(g.get_yticklabels(),rotation=0)

plt.title("Correlation Matrix Heatmap of Base Classifier Predictions")
```
